# Olympic Games Exploration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
athletes = pd.read_csv('athlete_events.csv')
regions = pd.read_csv('noc_regions.csv')
```

```python
athletes_.head()
```

```python
athletes_ = athletes.merge(regions,how='left', on='NOC')
```

```python
athletes_.rename(columns={'region':'Region', 'notes':'Notes'})
```

```python
athletes_.describe()
```

```python
athletes_.isna().sum()
```

```python
athletes_.columns[athletes_.isna().any()]
```

```python
athletes_[athletes_['Team']=='India']
```

### Countries with the highest Medals

```python
top_count=athletes_.Team.value_counts().sort_values(ascending= False).head(5)
top_count
plt.figure(figsize=(12,5))
plt.title('Participation by country')
plt.xlabel('Team')
plt.ylabel('Number of Times')
sns.barplot(x=top_count.index,y=top_count.values,palette='Set2')
```

### Age Distribution of the Athletes

```python
plt.figure(figsize=(12,5))
plt.title('Age Distribution of Athletes')
plt.xlabel('Age')
plt.ylabel('Frequency Density')
sns.histplot(data =athletes_, x= 'Age', binwidth=2, color='lime'  )
```

### Winter Olympic sports

```python
winter_olympic_sports = athletes_[athletes_.Season=='Winter'].Sport.unique()
winter_olympic_sports
```

### Winter Olympic Medal distribution wrt Sports


```python
winter_olympic = athletes_[athletes_.Season=='Winter']
winter_olympic=winter_olympic[['Season', 'Sport']] 
winter_olympic=winter_olympic.Sport.value_counts().sort_values()
winter_olympic
plt.figure(figsize=(10,8))
plt.title('Winter Olympic Medal distribution wrt Sports')
plt.pie(winter_olympic.values, labels=winter_olympic.index, autopct='%1.1f%%', shadow=False)
```

### summer_olympic_sports

```python
summer_olympic_sports = athletes_[athletes_.Season=='Summer'].Sport.unique()
summer_olympic_sports
summer_olympic = athletes_[athletes_.Season=='Summer']
summer_olympic = summer_olympic[['Season','Sport']].Sport.value_counts().sort_values()
summer_olympic
plt.figure(figsize=(18,15))
plt.title('summer olympic sports')
plt.xticks(rotation=75)
plt.bar(summer_olympic.index,summer_olympic.values,)
```

### Gender Participating in history of Olympics

```python
gender_counts = athletes_.Sex.value_counts()
gender_counts
plt.figure(figsize=(10,8))
plt.title('Gender Participating in history of Olympics')
plt.pie(gender_counts.values, labels=['Male','Female'], autopct='%1.1f%%', shadow=False)
```

### Exploring Female Athletes

```python
athletes_sorted_year=athletes_.sort_values(by=['Year'])
athletes_female= athletes_sorted_year[athletes_sorted_year['Sex']=='F']
athletes_female= athletes_female[['Sex','Year']]
y=athletes_female.Year.value_counts().sort_values(ascending=True)
y = pd.DataFrame(y)
y=y.reset_index().rename(columns= {'index': 'Year', 'Year':'Count'})
athletes_female
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
plt.title('Female Winners over the years')
sns.countplot(data=athletes_female, x='Year', palette='Spectral')
```

```python
gold= athletes_[athletes.Medal=='Gold']
gold.dropna()
```

### Exploring athletes greater than 60


```python
gold_60= gold[gold.Age>60]

gold_60= gold_60[['Sport']]
gold_60=gold_60.Sport.value_counts()


fig=plt.figure(figsize=(10,8))
#fig.label('Distribution of ')
plt.title('Distribution of Athletes greater than 60 according to their sport')
plt.pie(gold_60.values, labels=gold_60.index, autopct='%1.1f%%', shadow=False)
```

### Highest Gold Medal Earners

```python
x = pd.DataFrame(gold.region.value_counts().head()).reset_index().rename(columns={'index':'Region', 'region':'Count'})



plt.figure(figsize=(8,8))
plt.xlabel('Country')
plt.ylabel('Number of Gold Medals')
plt.title('Gold Medals per Country (Top 5)')
plt.bar(x.Region, x.Count, color=['orange', 'red', 'green', 'blue', 'cyan'])
```

### Exploring Weight and Height distribution of Medalists

```python
athletes_wh= athletes[['Weight','Height','Sex']]
athletes_wh.dropna()
plt.figure(figsize=(12,10))
axis = sns.scatterplot(x='Height', y='Weight', data = athletes_wh, hue = 'Sex')
plt.title('Weight vs Height distribution of Olympic Medalists')
```
